```python
# 1000 tracks per genre, 114 genres total
all_genres = data['track_genre'].value_counts()
print(all_genres)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of the 'popularity' column
plt.figure(figsize=(8, 5))
data['popularity'].hist(bins=20, edgecolor='k')
plt.title('Distribution of Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

```python
total_tracks = len(data)

# Calculate the number of tracks with popularity < 5
low_popularity_tracks = len(data[data['popularity'] == 0])

# Calculate the percentage
percentage_low_popularity = (low_popularity_tracks / total_tracks) * 100

print("Tracks with popularity == 0:", round(percentage_low_popularity, 2))
```

```python
import matplotlib.pyplot as plt

# Calculate percentage of tracks for each popularity value
popularity_distribution = data['popularity'].value_counts(normalize=True) * 100

# Sort the distribution by popularity values
popularity_distribution = popularity_distribution.sort_index()

plt.figure(figsize=(12, 6))
plt.bar(popularity_distribution.index, popularity_distribution.values)
plt.title('Percentage of Tracks by Popularity Rate')
plt.xlabel('Popularity (0-100)')
plt.ylabel('Percentage of Tracks')
plt.xticks(range(0, 101, 5))
plt.show()
```

```python
# Visualizing genre popularity
genre_popularity = data.groupby('track_genre')['popularity'].mean().sort_values()

print(genre_popularity.sort_values(ascending=False))

plt.figure(figsize=(16, 10))
genre_popularity.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Popularity for All Genres')
plt.xlabel('Genre')
plt.ylabel('Average Popularity')
plt.xticks(rotation=90)
plt.show()
```

```python
# Percentage of tracks with explicit value True or False
explicit_percentage = data['explicit'].value_counts(normalize=True) * 100

print(explicit_percentage)


# Count plot for 'explicit' column
sns.countplot(x='explicit', data=data)
plt.title('Count of Explicit Tracks')
plt.xlabel('Explicit (True/False)')
plt.ylabel('Count')
plt.show()
```

```python
# Visualizing duration_ms
plt.figure(figsize=(12, 6))
plt.hist(data['duration_ms'], bins=114, range=(0, 600000), edgecolor='black')
plt.title('Histogram of Track Durations (duration_ms)')
plt.xlabel('Duration (ms)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Shortest track
shortest_track = data[data['duration_ms'] == data['duration_ms'].min()]
# Longest track
longest_track = data[data['duration_ms'] == data['duration_ms'].max()]

shortest_track, longest_track
```

```python
# Find all tracks where the track name is missing (NaN)
tracks_missing_name = data[data['track_name'].isnull()]
tracks_missing_name
```

```python
# Tempo values
plt.figure(figsize=(12, 6))
plt.hist(data['tempo'], bins=114, edgecolor='black')
plt.title('Histogram of Track Tempo')
plt.xlabel('Tempo (BPM)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Tracks where tempo is 0
zero_tempo_tracks = data[data['tempo'] == 0]

zero_tempo_tracks
```
